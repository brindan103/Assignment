```
import matplotlib.pyplot as plt
import numpy as np
y=np.array([35,25,25,15])
mylabels=['Bus','car','plane','bike']
myexplode=[0,0.7,0,0.7]
plt.pie(y, labels=mylabels, explode=myexplode, shadow=True)
plt.legend(title='Vehicles-')
plt.show()
```
